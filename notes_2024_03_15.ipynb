## Systems of Differential Equations

The second order equation $ay'' + by' = cy = F(t)$ is equivalent to the first order system

\begin{align*}
u_0' &= u_1 \\
u_1' &= \frac{1}{a}(F(t) - bu_1 - cu_0)
\end{align*}

where $u_0 = y$ and $u_1 = y'$.

```python
a = 1; b = 1; c = 1;
F = lambda t: np.cos(t)
```

```python
f = lambda u,t: np.array([ u[1] , 1/a*(F(t) - b*u[1] - c*u[0]) ])
```

```python
f([0,1],0)
```

```python
u0 = [0,0]
t = np.linspace(0,50,1001)
u = spi.odeint(f,u0,t)
```
