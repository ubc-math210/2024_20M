## Plotting Solutions of ODEs

### First Order

Consider $y'= -y^2 t$ and plot the solution for $y(0)=0,...,1.9,2.0$.


\begin{align*}
y' &= -y^2 t \\
\int \frac{dy}{y^2} &= -\int t dt \\
\frac{-1}{y} &= -\frac{t^2}{2} + C \\
y &= \frac{1}{\frac{t^2}{2} + C} \\
\end{align*}

```python
t = np.linspace(0,5,100)
for y0 in np.arange(0.1,4,0.1):
    C = 1/y0
    y = 1/(t**2/2 + C)
    plt.plot(t,y,'k')
plt.grid(True)
plt.show()
```

### Second Order

Consider the equation $y'' + 2y' + 2y = 0$, $y(0)=0$, $y'(0)=c$. Plot the solution for $c=0,\dots,10$.

```python
# p(s) = as^2 + bs + c
a = 1
b = 2
c = 2
r1 = (-b + (b**2 - 4*a*c)**0.5)/(2*a)
r2 = (-b - (b**2 - 4*a*c)**0.5)/(2*a)
```

```python
r1
```

```python
r2
```

```python
alpha = r2.real
beta = r2.imag
```

```python
alpha,beta
```

```python
plt.figure(figsize=(7,3))
y = lambda t,v0: (v0/beta)*np.exp(alpha*t)*np.sin(beta*t)
t = np.linspace(0,5,500)
for v0 in np.arange(0,10,1):
    plt.plot(t,y(t,v0),'C0')
plt.grid(True)
plt.show()
```

## Parametric Curves

A parametric curve in 2D is of the form $\mathbf{x}(t) = (x_1(t),x_2(t))$. For example, the simplest parametric curve is the circle:

$$
\begin{align*}
x(t) &= \cos(t) \\
y(t) &= \sin(t)
\end{align*}
$$

```python
t = np.linspace(0,2*np.pi,50)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]), plt.grid(True)
plt.show()
```

```python
t = np.linspace(0,2*np.pi,100)
a = 2
x = np.cos(t)
y = np.sin(a*t)
plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]), plt.grid(True)
plt.show()
```

```python
t = np.linspace(0,12*np.pi,2000)
x = np.sin(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5)
y = np.cos(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5)
plt.figure(figsize=(3,3))
plt.plot(x,y,alpha=0.7)
plt.show()
```
