```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
N = 200
x = 1.5*np.random.rand(N)
y = x*(1 - x) + 0.02*np.random.randn(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
X = np.column_stack([x**0,x**1,x**2])
c = la.solve(X.T@X,X.T@y)
print(c)
```

```python
c.round(2)
```
