```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

## Polynomial Interpolation

```python
x = np.arange(0,5).astype(float)
y = [1.,4.,-2.,3.,2.]

X = np.vander(x,increasing=True)
# X = np.column_stack([x**n for n in range(len(x))])

c = la.solve(X,y)

xs = np.linspace(0,4,50)
ys = sum([c[n]*xs**n for n in range(len(c))])

plt.plot(xs,ys)
plt.plot(x,y,'.',ms=10)
plt.show()
```

```python
x = [-2.,-1.,0.,1.,2.]
y = [0.,0.,1.,0.,0.]

X = np.vander(x,increasing=True)
c = la.solve(X,y)
print(c)
xs = np.linspace(-2,2,50)
ys = sum([c[n]*xs**n for n in range(len(c))])

plt.plot(xs,ys)
plt.plot(x,y,'.',ms=10)
plt.show()
```
